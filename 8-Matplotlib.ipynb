Matplotlib is a Python library used for creating static, interactive, and animated visualizations. It is a versatile tool that enables users to generate a wide variety of plots, including line plots, scatter plots, bar charts, histograms, and more.

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

Types of Data Visualizations
Bar Chart- Rectangular bars or columns are used in bar charts to depict data.They can be used to highlight discrete categories or show changes over time. Horizontal bar charts and vertical bar charts are examples of different types.
Line Chart- Data is shown in line charts as a collection of data points connected by lines. They are perfect for displaying trends and patterns along an ongoing axis, like time.
Pie Chart- Pie charts show data as slices of a circle, each slice representing the percentage of each category. They are useful for showing parts of a whole or contrasting relative sizes.
Scatter Plot- Individual data points are displayed as dots on a two-dimensional grid in scatter plots. They are used to identify correlations and illustrate the link between two variables.
Histogram- The frequency or distribution of the data within each bin is shown by a histogram, which groups the data into bins. For displaying the distribution of continuous data, they work well.
Box Plot (Box-and-Whisker Plot)- Box plots use quartiles, outliers, and a box to represent the data distribution. They are useful for illustrating the skewness and distribution of data.

Bar Chart
Bar Chart- Rectangular bars or columns are used in bar charts to depict data.They can be used to highlight discrete categories or show changes over time. Horizontal bar charts and vertical bar charts are examples of different types.

When to use:

To compare categories or discrete groups.

For frequencies, counts, or averages across categories.

Example: Sales in different regions, exam scores by subject.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7 ]
y = [4, 3, 2, 5, 6, 7, 4]
plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

#Data
names = ['Aman', 'Pratiksha', 'Zeehan', 'Denim']
salaries = [20000, 15000, 80000, 50000]

# Create a bar chart
plt.bar(names, salaries, color = 'skyblue')


# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.grid()

# show the plot
plt.show()
```

Pie Chart
Pie charts show data as slices of a circle, each slice representing the percentage of each category. They are useful for showing parts of a whole or contrasting relative sizes.

When to use:

To show proportions/percentages of categories in a whole.

Best for small number of categories.

Example: Market share of companies, percentage of expenses in a budget.

```python
# Pie Chart

import matplotlib.pyplot as plt
x = ['Python', 'Java', 'C++', 'JavaScript']
y = [40, 30, 20, 10]
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# create a pie chart
plt.pie(sizes, labels = labels, colors =  colors, autopct = '%1.1f%%', startangle = 90, counterclock = False)

# Add title
plt.title("Pie Chart Example")

# show the plot
plt.show()
```

Scatter Plot
Scatter Plot- Individual data points are displayed as dots on a two-dimensional grid in scatter plots. They are used to identify correlations and illustrate the link between two variables.

When to use:

To visualize the relationship/correlation between two continuous variables.

Detects patterns, clusters, or outliers.

Example: Height vs weight of students.

```python
x = [3, 5, 8, 0, 3, 7, 1, 4]
y = [2, 4, 8, 5, 3, 2, 9, 4]
plt.scatter(x, y)

plt.show()
```

```python
age = [30, 15, 30, 20, 13, 17, 10, 40]
Income = [2000, 4000, 80000, 55000, 3300, 21000, 90000, 44000]

#plot the scatter plot
plt.scatter(age, Income, marker = '*', color = 'red')
#plt.grid()
plt.grid(alpha = 0.2)
# plot the label and title
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Scatter Chart Example")

# show the plot
plt.show()
```

Histogram
Histogram- The frequency or distribution of the data within each bin is shown by a histogram, which groups the data into bins. For displaying the distribution of continuous data, they work well.

When to use:

To show the distribution of a single continuous variable.

Useful for seeing frequency, spread, skewness, and shape of data (normal, uniform, etc.).

Example: Age distribution of employees, exam scores of students.

```python
l = [1,2,2,3,3,4,4,4,4,4,4,5,6,7,7,7,7,7,7,7,7]
plt.hist(l)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generate random data
data = np.random.randint(1000, size = 500)  # size is the no. of samples

# create histogram
plt.hist(data, bins=30, color = 'green', edgecolor = 'black')

# Add labels and titles
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# show the plot
plt.show()
```

Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading a Dataset
Seaborn comes with built-in datasets that you can use for practise. Let's load the "tips" dataset

```python
# load the tips dataset

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips.columns
```

```python
sns.scatterplot(x ="total_bill", y = "tip", data = tips)
plt.title("Total bill vs Tips")
plt.grid()
plt.show()
```

```python
# Line Plot
sns.lineplot(x ="total_bill", y = "tip", data = tips)
plt.title("Total bill vs Tips")
plt.grid()
plt.show()
```

```python
# Bar Plot
sns.barplot(x = "day", y ="total_bill", data = tips)
plt.title("Bar Plot day vs total_bill")
plt.show()
```

KDE is like a smoothed version of a histogram.

Instead of putting data into rigid bins, KDE places a small bump (a kernel function, usually Gaussian bell curve) at each data point.

Then it adds up all those bumps to get a smooth curve that estimates the probability density function (PDF) of the data.

Box Plot
A box plot (also called box-and-whisker plot) is a graph that shows the spread and distribution of a dataset using five key summary values.

```python
sns.boxplot(x = "day", y= "total_bill", data = tips)
plt.title("Total Bill by Day(Box Plot)")
plt.show()
```

Pair Plot
A pairplot is a powerful visualization tool that shows pairwise relationships between multiple columns in a dataset  especially useful for exploring numerical data.

A pairplot creates:

Scatter plots between every pair of numerical columns

Histograms (or KDEs) on the diagonal for individual columns

```python
sns.pairplot(tips, hue = "sex")

plt.show()
```

Heat Map
A heatmap is a data visualization technique that uses color to represent the magnitude of values in a matrix or 2D data. Or you can say a heatmap is like a colorful table  instead of just showing numbers, it uses colors to make patterns and trends easy to see.

```python
# selected the numeric column
selected_columns = ['total_bill', 'tip', 'size']

corr = tips[selected_columns].corr() # checking for the correlation -- it mainly works with numerical data

# Plot heatmap
sns.heatmap(corr,annot = True, cmap='YlGnBu')
plt.title("Correlation Heatmap (Selected Columns)")
plt.show()
```

Violin Plot
Violin Plot is a method to visualize the distribution of numerical data of different variables. It is quite similar to Box Plot but with a rotated plot on each side, giving more information about the density estimate on the y-axis.

A violin plot consists of four components.

A white Centered Dot at the middle of the graph - The white dot point at the middle is the median of the distribution.
A thin gray bar inside the plot - The bar in the plot represents the Quartile range of the distribution.
A long thin line coming outside from the bar - The thin line represents the rest of the distribution which is calculated by the formulae Q1-1.5 IQR for the lower range and Q3+1.5 IQR for the upper range. The point lying beyond this line are considered as outliers.
A line boundary separating the plot- A KDE plot is used for defining the boundary of the violin plot it represents the distribution of data points.

```python
import seaborn as sns
# Create a violin plot
sns.violinplot(x='day', y='tip', data=tips)

# Add title and show the plot
plt.title('Violin Plot of Tips by Day')
plt.show()
```
