Python Object
Objects are class instances which are used to access methods and data (defined by their class) for modifying their state.

Syntax: obj = ClassName()

Self
Class methods must have an extra first parameter in the method definition. We do not give a value for this parameter when we call the method, Python provides it.
It is denoted by keyword “self”.
If we have a method that takes no arguments, we still have one argument.
When we call a method of this object as myobject.method(arg1, arg2), this is automatically converted by Python into MyClass.method(myobject, arg1, arg2).
It is used to access data members (attributes).
self parameter is a reference to the current instance of the class. It allows us to access the attributes and methods of the object.

Init
init method is the constructor in Python, automatically called when a new object is created. It initializes the attributes of the class.

```python
# define a simple class

class Student:
  def __init__(self, name, course):
    # constructor
    self.name = name # the self parameter is a reference to the current instance of the class, and is used to access variaables that belong to the class
    self.course = course

  def display(self):
    print(f"{self.name} is enrolled in {self.course}")
```

```python
# create objects(instance)
s1 = Student("Alice", "Data Science")
s2 = Student("Bob", "Machine Learning")
```

```python
# call a method on objects
s1.display()
s2.display()
```

Polymorphism
Polymorphism (Same function, different behavior)

It refers to the ability to use the same function or method name, but with different behavior depending on the object it's acting on.

You can write functions or logic that work on different kinds of objects, without knowing their specific class, as long as they all have a common method or behavior.

```python
print(len("Hello"))  # string length
print(len([1,2,3])) # list length

print(max(1, 3, 2))  # Maximum of integers
print(max("a", "z", "m"))  # Maximum in strings
```

```python
class Circle:
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius  # pi*r^2

class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self. width = width

  def area(self):
    return self.length * self.width

class Square:
  def __init__(self, side):
    self.side = side

  def area(self):
    return self.side * self.side


# Polymorphic function

def print_area(shape):
  print("Area is", shape.area())

# different objects with different __init__ parameters
c = Circle(5)
r = Rectangle(4, 6)
s = Square(3)

# use polymorphism

for shape in [c, r, s]:
  print_area(shape)
```
